## Continuous latent space models for network data
Lets first recall few elements about what is a CLS (continuous latent space model) in the case of graph data according to A. L. Smith and all:
A fully parametric model for network data provides a natural way to think about utilizing a lower-dimensional latent space for the nodes in the network. A generic form for such a model is given below:

\begin{align}
  Y_{i,j} &\sim Bernouilli(p_{i,j}), \quad & i  \neq j; \; i,j = 1,\dots,n \\
  logit(p_{ij}) &= \alpha + s(z_i, z_j) \\
  p_{ij} &= logistic \left( \alpha + s(z_i, z_j) \right) \\
  z_i \in \mathbb{R}^t, z_i &\sim f_t(z|\psi) \quad & i=1,\dots,n,
\end{align}

where $Y_{i,j}$ is a binary indicator of a tie between nodes i and j, $z_i$ is the $i^{th}$ nodeâ€™s t-dimensional unobserved latent vector in the t-dimensional space $\mathbb{R}^t$, and $\psi$ parameterizes the distribution of the latent vectors in $\mathbb{R}^t$. The generic latent space.

We also recall that:
* $$ logit(p) = logistic^{-1}(p) = ln \left( \frac{p}{1-p} \right) \quad {\text{for}} \quad p\in ]0,1[ $$
* $$ logistic(\alpha) = \frac{1}{1+\exp(-\alpha)} $$

```python
# Just remember logistic and logit
fig, ax = plt.subplots(1,2, figsize=(12,6))
eps = np.finfo(np.float32).eps

# Logit is defined on (0, 1)
lit_x = np.linspace(0+eps,1-eps,1024)
logit_func = np.log(lit_x/(1-lit_x))

# Logit is defined on R
lic_x = np.linspace(-5,5,4096)
logistic_func = 1/(1+np.exp(-lic_x))

fig.suptitle("Logit and logistic functions")
ax[0].set_title("Logit")
ax[0].plot(lit_x, logit_func)
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

ax[1].set_title("Logistic")
ax[1].plot(lic_x, logistic_func)
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")
```
